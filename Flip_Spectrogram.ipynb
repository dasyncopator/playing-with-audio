```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import librosa
import soundfile as sf
```

```python
#Load audio
filename = r'C:\Users\Tanay Mannikar\Music\Da Syncopator\Released\Stellar Tour\Galaxy Vibes.wav'
y, sr = librosa.load(filename, sr=44100)

# Perform STFT to get spectrogram data
audio = y
audio_stft = np.abs(librosa.stft(audio))
```

```python
# Plot spectrogram, dB frequency scale
df = pd.DataFrame(data=audio_stft)
plt.specgram(audio)
plt.show()
```

```python
# Invert frequency bank rows to flip spectrogram
df_flip = df.T

# Perform ISTFT to generate new audio
audio_hat = librosa.istft(df_flip.values)
```

```python
# Plot new spectrogram
plt.specgram(audio_hat)
plt.show()
```

```python
# Generate .wav file of resultant audio
filename_flip = r'C:\Users\Tanay Mannikar\Documents\Inverberator\audio\galaxy_vibes_flipped.wav'
sf.write(filename_flip, audio_hat, sr, 'PCM_16')
```
